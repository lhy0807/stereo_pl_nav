```python
import os
```

```python
isec1_voxel_dir = "./ISEC1/voxel"
isec1_left_dir = "./ISEC1/left"
isec1_right_dir = "./ISEC1/right"

disp_names = []
for root, dirs, files in os.walk(isec1_voxel_dir):
    for filename in files:
#         print(os.path.join(root, filename))
        disp_names.append(filename)
```

```python
len(disp_names)
```

```python
disp_names
```

```python
with open('ISEC1_train.txt', 'w') as f:
    for name in disp_names[:-200]:
        line = os.path.join(isec1_left_dir, name) + " " + os.path.join(isec1_right_dir, name) + " " + \
                os.path.join(isec1_voxel_dir, name) + "\n"
        f.writelines(line)
```

```python
with open('ISEC1_test.txt', 'w') as f:
    for name in disp_names[-200:]:
        line = os.path.join(isec1_left_dir, name) + " " + os.path.join(isec1_right_dir, name) + " " + \
                os.path.join(isec1_voxel_dir, name) + "\n"
        f.writelines(line)
```
