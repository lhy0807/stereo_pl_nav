```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating vectors X and Y
d = np.linspace(1, 200, 200)
z = 540 / d

 
fig = plt.figure(figsize = (5, 5))
# Create the plot
plt.plot(d, z)
for d_i in np.arange(1, 200, 10):
    plt.vlines(x=d_i, ymin=-10, ymax=540/d_i, colors="y")
    plt.hlines(y=540/d_i, xmin=0, xmax=d_i, colors="g", linestyles="--")
plt.xlabel("Disparity")
plt.ylabel("Depth")

# Show the plot
plt.savefig("origin_func.png")
plt.show()
```

```python
fig = plt.figure(figsize = (5, 5))
plt.hist(z[z < 200], bins=10)
plt.xlabel("Depth")
plt.ylabel("Count")
plt.savefig("origin_hist.png")
plt.show()
```

```python
# Creating vectors X and Y
z = np.linspace(4, 84, 84)
d = 540 / z

 
fig = plt.figure(figsize = (5, 5))
# Create the plot
plt.plot(d, z)
for z_i in np.arange(4, 84, 4):
    plt.hlines(y=z_i, xmin=-1, xmax=540/z_i, colors="g", linestyles="--")
    plt.vlines(x=540/z_i, ymin=0, ymax=z_i, colors="y")
plt.xlabel("Disparity")
plt.ylabel("Depth")

# Show the plot
plt.savefig("our_func.png")
plt.show()
```

```python
fig = plt.figure(figsize = (5, 5))
plt.hist(z[z < 200], bins=10)
plt.xlabel("Depth")
plt.ylabel("Count")

plt.savefig("our_hist.png")
plt.show()
```
